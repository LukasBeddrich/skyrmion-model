```python
### excistation funktion ###
def gauss(x, x0, sig, amp):
    """
    x0: excitation energy
    sig: linewidth
    amp: weight --> relative ampl?
    """
    norm = (sqrt(2.*pi) * sig)
    return amp * exp(-0.5*((x-x0)/sig)**2.) / norm
```
