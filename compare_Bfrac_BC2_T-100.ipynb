```python
### excistation funktion ###
def gauss(x, x0, sig, amp):
    """
    x0: excitation energy
    sig: linewidth
    amp: weight --> relative ampl?
    """
    norm = (sqrt(2.*pi) * sig)
    return amp * exp(-0.5*((x-x0)/sig)**2.) / norm
```

```python
#### SQlite connection ###
conn = smr.create_EW_connection()
cur = conn.cursor()
```

```python
#### load Bfrac dep at one Kvector ####
cur.execute('SELECT Energy, Weight FROM R_3 WHERE Bfrac == 0.2')
data = asarray(cur.fetchall())
```

```python
khneg10, khneg05, kh0, khpos05, khpos10, khpos15, khpos20 = [data[i*14:(i+1)*14] for i in xrange(7)]
```

```python
### spectrum for Bfrac = 0.2 and q = 0 kh ###
energy = np.linspace(0.,80., 241)
khneg10_sqw_sing = [gauss(energy, khneg10[j][0], 14.247, khneg10[j][1]) for j in xrange(len(khneg10))]
khneg10_sqw = np.sum(khneg10_sqw_sing, axis = 0)
```

```python
figneg10 = figure(figsize = (8.,6.))
plt.plot(energy, khneg10_sqw, color = 'b', ls = '-', lw = 2., label = r'|q| $\approx$ 0 kh')
for i in khneg10_sqw_sing:
    plot(energy, i, color = 'steelblue', ls = '--', lw = 1.5)
legend(loc = 'best', fontsize = 14.)
show()
```

```python
#### spectrum for Bfrac = 0.2 and q = 0.5 kh ###
khneg05_sqw_sing = [gauss(energy, khneg05[j,0], 14.247, khneg05[j,1]) for j in xrange(len(khneg05))]
khneg05_sqw = np.sum(khneg05_sqw_sing, axis = 0)
#### spectrum for Bfrac = 0.2 and q = 1.0 kh ###
kh0_sqw_sing = [gauss(energy, kh0[j,0], 14.247, kh0[j,1]) for j in xrange(len(kh0))]
kh0_sqw = np.sum(kh0_sqw_sing, axis = 0)
#### spectrum for Bfrac = 0.2 and q = 1.5 kh ###
khpos05_sqw_sing = [gauss(energy, khpos05[j,0], 14.247, khpos05[j,1]) for j in xrange(len(khpos05))]
khpos05_sqw = np.sum(khpos05_sqw_sing, axis = 0)
#### spectrum for Bfrac = 0.2 and q = 2.0 kh ###
khpos10_sqw_sing = [gauss(energy, khpos10[j,0], 14.247, khpos10[j,1]) for j in xrange(len(khpos10))]
khpos10_sqw = np.sum(khpos10_sqw_sing, axis = 0)
#### spectrum for Bfrac = 0.2 and q = 2.5 kh ###
khpos15_sqw_sing = [gauss(energy, khpos15[j,0], 14.247, khpos15[j,1]) for j in xrange(len(khpos15))]
khpos15_sqw = np.sum(khpos15_sqw_sing, axis = 0)
#### spectrum for Bfrac = 0.2 and q = 3.0 kh ###
khpos20_sqw_sing = [gauss(energy, khpos20[j,0], 14.247, khpos20[j,1]) for j in xrange(len(khpos20))]
khpos20_sqw = np.sum(khpos20_sqw_sing, axis = 0)
```

```python
fig2 = figure(figsize = (12.,9))
title(r'Inelastic Scattering in A-Phase of MnSi', fontsize = 21.)
plot(energy, khneg10_sqw, 'b-', label = r'|q| $\approx$ 0 kh', lw = 2.)
plot(energy, khneg05_sqw, 'r-', label = r'|q| $\approx$ 0.5 kh', lw = 2.)
plot(energy, kh0_sqw, 'g-', label = r'|q| $\approx$ 1 kh', lw = 2.)
plot(energy, khpos05_sqw, 'y-', label = r'|q| $\approx$ 1.5 kh', lw = 2.)
plot(energy, khpos10_sqw, 'c-', label = r'|q| $\approx$ 2 kh', lw = 2.)
plot(energy, khpos15_sqw, color = 'fuchsia', ls = '-', label = r'|q| $\approx$ 2.5 kh', lw = 2.)
plot(energy, khpos20_sqw, 'k-', label = r'|q| $\approx$ 3 kh', lw = 2.)
legend(loc = "best", fontsize = 14.)
#xlim(xmax = 70.)
ylim(ymax = .09)
xlabel('Energy [arb.u.]', fontsize = 16.)
ylabel('Intensity [arb.u.]', fontsize = 16.)
tick_params('both', labelsize = 14.)
show()
```
